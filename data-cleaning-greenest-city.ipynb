```python
import requests
import pandas as pd

url = "https://opendata.vancouver.ca/api/explore/v2.1/catalog/datasets/greenest-city-projects/records"

params = {"limit": 100, "offset": 0}
all_records = []

while True:
    response = requests.get(url, params=params)
    records = response.json().get("results", [])
    all_records.extend(records)
    if len(records) < 100:
        break
    params["offset"] += 100

greenest_city_projects = pd.DataFrame(all_records)
greenest_city_projects.to_csv("greenest_city_projects.csv", index=False)
```

```python
import ast

greenest_city_projects = pd.read_csv("greenest_city_projects.csv")

def parse_coords_gcp(geom_str):
    try:
        geom = ast.literal_eval(geom_str)
        coords = geom["geometry"]["coordinates"]
        return pd.Series({"longitude": coords[0], "latitude": coords[1]})
    except:
        return pd.Series({"longitude": None, "latitude": None})

greenest_city_projects[["longitude", "latitude"]] = greenest_city_projects["geom"].apply(parse_coords_gcp)
greenest_city_projects.to_csv("greenest_city_projects.csv", index=False)
```
